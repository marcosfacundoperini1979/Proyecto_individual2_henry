por ultimo vamos a trabajar con la columna x e y para un visualizacion correcta a posteriori

```python
# Filtrar filas con posiciones no válidas
valid_positions = ~(dataset_homicidios['pos x'].str.contains('[^0-9.,-]') | dataset_homicidios['pos y'].str.contains('[^0-9.,-]') | (dataset_homicidios['pos x'] == '.') | (dataset_homicidios['pos y'] == '.'))

# Establecer posiciones no válidas en NaN
dataset_homicidios.loc[~valid_positions, ['pos x', 'pos y']] = np.nan

# Convertir las cadenas a tipo numérico (float)
dataset_homicidios['pos x'] = dataset_homicidios['pos x'].str.replace(',', '.').astype(float)
dataset_homicidios['pos y'] = dataset_homicidios['pos y'].str.replace(',', '.').astype(float)
```

Columna Fallecimiento  se pasa a formato datetime  luego a date eliminando las horas ya que no las usare  y poor ultimo  relleno con vacio los NaT  que me produjo el coerce

```python
dataset_victimas["FECHA_FALLECIMIENTO"] = pd.to_datetime(dataset_victimas["FECHA_FALLECIMIENTO"], format='%Y-%m-%d', errors='coerce')
dataset_victimas["FECHA_FALLECIMIENTO"] = dataset_victimas["FECHA_FALLECIMIENTO"].dt.date
dataset_victimas["FECHA_FALLECIMIENTO"] = dataset_victimas ["FECHA_FALLECIMIENTO"].replace(pd.NaT,"")
```

trabajo sobre la columna EDAD para luego hacer un grupo Etario

```python
dataset_victimas ["EDAD"] = pd.to_numeric(dataset_victimas["EDAD"], errors="coerce")
dataset_victimas ["EDAD"] = dataset_victimas["EDAD"].astype("Int64")
dataset_victimas["EDAD"] = dataset_victimas["EDAD"].replace([" ","SD"], np.nan)
dataset_victimas["EDAD"] = dataset_victimas["EDAD"].fillna(0.0)
dataset_victimas ["EDAD"].unique()
```

Creo Rango Etario

```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['Menos de 20', ' Entre 21- 40', 'Entre 41-60', 'Entre 61-80', 'Más de 80']
dataset_victimas['Rango_etario'] = pd.cut(dataset_victimas['EDAD'], bins=bins, labels=labels, right=False)
dataset_victimas.head(5)
```
